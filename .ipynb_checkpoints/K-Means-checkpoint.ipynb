# Data koordinat letak lokasi

```python
# Data Koordinat kota pada peta
data=[[34,32], # Lokasi 0
      [90,87], # Lokasi 1
      [35,98], # Lokasi 2
      [24,24], # Dan seterusnya...
      [90,0],
      [12,12],
      [8,23],
      [12,90],
      [33,43],
      [64,81],
      [39,29],
      [15,19]]

# 3 titik cluster ditentukan
K=[[12,90], # Titik cluster A
   [17,12], # Titik cluster B
   [12,24]] # Titik cluster C
```

# Menghitung Jarak antara koordinat kota dan titik cluster

```python
  print ('             K0        K1        K2 ')
# Jarak lokasi 0 dengan titik cluster
jarak0=((data[0][0]-K[0][0])**2+(data[0][1]-K[0][1])**2)**(1/2)
jarak1=((data[0][0]-K[1][0])**2+(data[0][1]-K[1][1])**2)**(1/2)
jarak2=((data[0][0]-K[2][0])**2+(data[0][1]-K[2][1])**2)**(1/2)
print('lokasi 0 : {:5.2f}     {:5.2f}      {:5.2f}'.format(jarak0,jarak1,jarak2))

# Jarak kota 1 dengan titik cluster
jarak0=((data[1][0]-K[0][0])**2+(data[1][1]-K[0][1])**2)**(1/2)
jarak2=((data[1][0]-K[1][0])**2+(data[1][1]-K[1][1])**2)**(1/2)
jarak2=((data[1][0]-K[2][0])**2+(data[1][1]-K[2][1])**2)**(1/2)
print('lokasi 1 : {:5.2f}     {:5.2f}      {:5.2f}'.format(jarak0,jarak1,jarak2))

# Jarak kota 2 dengan titik cluster
jarak0=((data[2][0]-K[0][0])**2+(data[2][1]-K[0][1])**2)**(1/2)
jarak1=((data[2][0]-K[1][0])**2+(data[2][1]-K[1][1])**2)**(1/2)
jarak2=((data[2][0]-K[2][0])**2+(data[2][1]-K[2][1])**2)**(1/2)
print('lokasi 2 : {:5.2f}     {:5.2f}      {:5.2f}'.format(jarak0,jarak1,jarak2))

# Jarak kota 3 dengan titik cluster
jarak0=((data[3][0]-K[0][0])**2+(data[3][1]-K[0][1])**2)**(1/2)
jarak1=((data[3][0]-K[1][0])**2+(data[3][1]-K[1][1])**2)**(1/2)
jarak2=((data[3][0]-K[2][0])**2+(data[3][1]-K[2][1])**2)**(1/2)
print('lokasi 3 : {:5.2f}     {:5.2f}      {:5.2f}'.format(jarak0,jarak1,jarak2))

# Dan Seterusnya...
print('dan seterusnya... # Hasil dibulatkan ke 2 angka dibelakang koma')
```

```python
# Jarak antara lokasi dan titik cluster (lebih simple)
jarak=[]
for i in range (len(data)):
    subjarak=[]
    for j in range (len(K)):
        h=((data[i][0]-K[j][0])**2+(data[i][1]-K[j][1])**2)**(1/2)
        subjarak.append(h)
    jarak.append(subjarak)
jarak
```

```python
# Program untuk cluster
clusterA=[]
for i in range (len(data)):
    jarak=[]
    for j in range (len(K)):
        h=((data[i][0]-K[j][0])**2+(data[i][1]-K[j][1])**2)**(1/2)
        jarak.append(h)
    for k in range (len(jarak)):
        if min(jarak)==jarak[k]:
            if k==0:
                clusterA.append(data[i])
clusterB=[]
for i in range (len(data)):
    jarak=[]
    for j in range (len(K)):
        h=((data[i][0]-K[j][0])**2+(data[i][1]-K[j][1])**2)**(1/2)
        jarak.append(h)
    for k in range (len(jarak)):
        if min(jarak)==jarak[k]:
            if k==1:
                clusterB.append(data[i])
clusterC=[]
for i in range (len(data)):
    jarak=[]
    for j in range (len(K)):
        h=((data[i][0]-K[j][0])**2+(data[i][1]-K[j][1])**2)**(1/2)
        jarak.append(h)
    for k in range (len(jarak)):
        if min(jarak)==jarak[k]:
            if k==2:
                clusterC.append(data[i])
```

```python
print(clusterA)
print(clusterB)
print(clusterC)
```

```python
# Progam cluster (lebih simple)
cluster=[]
for i in range (len(K)):
    clustertemp=[]
    for j in range (len(data)):
        jarak=[]
        for k in range (len(K)):
            h=((data[j][0]-K[k][0])**2+(data[j][1]-K[k][1])**2)**(1/2)
            jarak.append(h)
        for l in range (len(jarak)):
            if min(jarak)==jarak[l] and l==i:
                clustertemp.append(data[j])
    cluster.append(clustertemp)
print(cluster)
```

```python
# Program mengganti K baru
for i in range (2):
    jumlah=0
    for j in range (len(clusterA)):
        jumlah+=clusterA[j][i]
    rata=jumlah/len(clusterA)
    K[0][i]=rata
for i in range (2):
    jumlah=0
    for j in range (len(clusterB)):
        jumlah+=clusterB[j][i]
    rata=jumlah/len(clusterB)
    K[1][i]=rata
for i in range (2):
    jumlah=0
    for j in range (len(clusterC)):
        jumlah+=clusterC[j][i]
    rata=jumlah/len(clusterC)
    K[2][i]=rata
print ('K baru :',K)
```

```python
# Program mengganti K baru yang (lebih simple)
for i in range (len(cluster)):
    for j in range (2):
        jumlah=0
        for k in range (len(cluster[i])):
            jumlah+=cluster[i][k][j]
        rata=jumlah/len(cluster[i])
        K[i][j]=rata
print ('K baru :',K)
```

```python
def kooracak(n,minimal,maximal):
    from random import randint
    data=[]
    for i in range (n):
        data2=[]
        for j in range (2):
            a=randint(minimal,maximal)
            data2.append(a)
        data.append(data2)
    return data


kooracak(10,0,999)
```

# Proses Clustering dengan K-Means dengan data random

```python
# Data Koordinat lokasi pada peta
data=kooracak(500,0,10000)

# 3 titik cluster ditentukan
K=kooracak(5,0,10000)
```

```python
K1=[]
K2=[1]

while K2!=K1:
    cluster=[]
    for i in range (len(K)):
        clustertemp=[]
        for j in range (len(data)):
            jarak=[]
            for k in range (len(K)):
                h=((data[j][0]-K[k][0])**2+(data[j][1]-K[k][1])**2)**(1/2)
                jarak.append(h)
            for l in range (len(jarak)):
                if min(jarak)==jarak[l] and l==i:
                    clustertemp.append(data[j])
        cluster.append(clustertemp)
    for i in range (len(cluster)):
        for j in range (2):
            jumlah=0
            for k in range (len(cluster[i])):
                jumlah+=cluster[i][k][j]
            rata=jumlah/len(cluster[i])
            K[i][j]=rata
    K2=K1
    K1=K
print ('Centroid :',K)
print ('Cluster :',cluster)
```

```python
def titik(n,warna,nama):
    clusterx=[]
    clustery=[]
    for i in range(len(cluster[n])):
        clusterx.append(cluster[n][i][0])
        clustery.append(cluster[n][i][1])
    plt.scatter(clusterx,clustery, s = 30, c = warna, label = nama)
def plotcentroid(warna,nama):
    centx=[]
    centy=[]
    for i in range (len(K)):
        centx.append(K[i][0])
        centy.append(K[i][1])
    plt.scatter(centx,centy, s = 100, c = warna, label = nama)
```

```python
titik(0,'red','Cluster 1')
titik(1,'blue','Cluster 2')
titik(2,'green','Cluster 3')
titik(3,'yellow','Cluster 4')
titik(4,'cyan','Cluster 5')
plotcentroid('black','Centroid')

#plt.title('Cluster titik lokasi')
#plt.xlabel('koordinat x')
#plt.ylabel('koordinat y')
plt.legend()
plt.show()
```
